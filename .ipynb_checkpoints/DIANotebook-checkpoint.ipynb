```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pylab as py
from cvxpy import *
from dia3 import dia
from dia3 import dialin

imageRef = cv2.imread('createImage/doubleOut/out1.png',0) #READ IN WITH CV2
(m,n) = imageRef.shape

image2 = cv2.imread('createImage/doubleOut/out15.png',0) #READ IN WITH CV2

kb = py.ones((3,3)) # DEFINE THE KERNEL BASIS
#print(kb)
kb=[[0,0,0],[0,1,0],[0,0,0]]
# Compute Difference Image Analysis:
#m, kern, bkg, chisq = dialin(imageRef, image2, kb,noback=True)
m, kern, bkg, chisq = dia(imageRef, image2, kb,noback=True)

# Display results:
py.figure()
py.subplot(231)
py.imshow(imageRef)
py.title('Reference image')
py.colorbar(orientation='horizontal', fraction=0.046, pad=0.04)
py.subplot(232)
py.imshow(image2)
py.title('Observed image')
py.colorbar(orientation='horizontal', fraction=0.046, pad=0.04)
py.subplot(234)
py.imshow(kern)
py.title('Optimal kernel')
py.colorbar(orientation='horizontal', fraction=0.046, pad=0.04)
py.subplot(235)
py.imshow(m)
py.title('Convolved Reference')
py.colorbar(orientation='horizontal', fraction=0.046, pad=0.04)
py.subplot(236)
py.imshow(image2 - m)
py.title('Residuals')
py.colorbar(orientation='horizontal', fraction=0.046, pad=0.04)

py.rcParams['figure.figsize'] = (100.0, 100.0)

#PLOT RESULTS
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 12))
ax1.imshow(imageRef, origin='lower', interpolation='nearest', cmap='Greys_r')
ax2.imshow(image2 - imageRef, origin='lower', interpolation='nearest', cmap='Greys_r')
plt.show()
```
